```python
from IPython.display import display, Markdown

import sys
sys.path.append('.')

import matplotlib.pyplot as plt

from pipeline import pipeline
```

```python
sinusoidal = pipeline.complex('sinusoidal')
```

```python
complex = sinusoidal.build()
```

```python
mg = complex.frames[0].to_graph()
mg_noisy = complex.frames[1].to_graph()
```

```python
samplings = [None, 5, 10, 20]

fig, axes = plt.subplots(len(samplings), 2, figsize=(2 * 12, len(samplings) * 12))

for ax_row, sampling in zip(axes, samplings):
  if sampling is not None:
    g = mg.sample(sampling)
    gn = mg_noisy.sample(sampling)
  else:
    g = mg
    gn = mg_noisy
    
  g.draw(ax=ax_row[0])
  gn.draw(ax=ax_row[1])
```
